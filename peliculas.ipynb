```python
import numpy as np

# 1. Nuestro catálogo con vectores: [Acción, Comedia, Drama, Sci-Fi]
catalogo = {
    "Mad Max": np.array([1, 0, 0, 0]),
    "Toy Story": np.array([0, 1, 0, 0]),
    "The Joker": np.array([0, 0, 1, 0]),
    "Interstellar": np.array([1, 0, 0, 1]),
    "Superbad": np.array([0, 1, 0, 0]),
    "Die Hard": np.array([1, 0, 0, 0]),
    "Parasite": np.array([0, 0, 1, 0]),
    "Inception": np.array([1, 0, 0, 1]),
    "Avengers": np.array([1, 0, 0, 1]),
    "The Hangover": np.array([0, 1, 0, 0]),
    "Titanic": np.array([0, 0, 1, 0])
}

# Creamos una lista con los nombres para poder acceder por índice (número)
nombres_peliculas = list(catalogo.keys())

def calcular_similitud(peli1, peli2):
    v1 = catalogo[peli1]
    v2 = catalogo[peli2]
    denominador = np.linalg.norm(v1) * np.linalg.norm(v2)
    if denominador == 0: return 0
    return np.dot(v1, v2) / denominador

# --- INICIO DEL PROGRAMA ---
print("--- 🎬 Recomendador por Números de helpluchi ---")
print("Selecciona tus películas por número:")

# Mostramos el menú numerado
for i, nombre in enumerate(nombres_peliculas):
    print(f"{i}. {nombre}")

print("-" * 40)

try:
    # Pregunta 1 por número
    idx1 = int(input("Pregunta 1: Elige el NÚMERO de la primera película: "))
    # Pregunta 2 por número
    idx2 = int(input("Pregunta 2: Elige el NÚMERO de la segunda película: "))

    # Obtenemos los nombres usando los números
    peli_a = nombres_peliculas[idx1]
    peli_b = nombres_peliculas[idx2]

    resultado = calcular_similitud(peli_a, peli_b)
    print(f"\n✅ Has comparado '{peli_a}' con '{peli_b}'")
    print(f"📊 La similitud es del {resultado * 100:.0f}%")

except (ValueError, IndexError):
    print("\n❌ Error: Por favor, introduce un número válido de la lista.")

# --- PARTE 2: BUSCADOR POR PALABRAS CLAVE ---
print("\n" + "-" * 40)
print("--- 🔍 Buscador Inteligente ---")
print("Categorías disponibles: 0:Acción, 1:Comedia, 2:Drama, 3:Sci-Fi")
opcion_cat = int(input("Elige el NÚMERO de la categoría que buscas: "))

categorias_nombres = ["Acción", "Comedia", "Drama", "Sci-Fi"]

if 0 <= opcion_cat <= 3:
    print(f"\nPelículas encontradas en '{categorias_nombres[opcion_cat]}':")
    for nombre, vector in catalogo.items():
        if vector[opcion_cat] == 1:
            print(f"-> {nombre}")
else:
    print("Número de categoría no válido.")

print("\n--- ¡Sigue así, dominas el código como un profesional! ---")
```
