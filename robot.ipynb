```python
import time
import random

class RobotPaseador:
    def __init__(self, nombre):
        self.nombre = nombre
        # Base de datos de rutas (Calle: Tipo de zona)
        self.db_rutas = {
            "Calle Roble": "Segura",
            "Avenida Parque": "Agradable",
            "Sendero Verde": "Segura"
        }
        self.rutas_programadas = []
        self.bolsitas = 5
        self.tanque_agua_ml = 1000  # 1 Litro
        self.tiempo_restante = 30    # Minutos de paseo

        print(f"--- 🤖 {self.nombre} INICIADO ---")

    def planificar_ruta(self):
        """Verifica si hay rutas en la memoria, si no, las configura."""
        if not self.rutas_programadas:
            print("🗂️ Memoria vacía. Accediendo a Base de Datos para configurar rutas...")
            # Simulamos que elige las rutas seguras y agradables de la DB
            self.rutas_programadas = list(self.db_rutas.keys())
            time.sleep(1)
            print(f"✅ Rutas configuradas: {self.rutas_programadas}")
        else:
            print(f"🚀 Iniciando paseo con rutas ya programadas: {self.rutas_programadas}")

    def sensores_limpieza(self):
        """Simula la detección de necesidades del perro."""
        # Genera un evento aleatorio: 0=nada, 1=popó, 2=pis, 3=charco
        return random.choice([0, 1, 2, 3, 0, 0])

    def ejecutar_paseo(self):
        self.planificar_ruta()

        for calle in self.rutas_programadas:
            # CONDICIÓN DE TIEMPO (Regreso a casa)
            if self.tiempo_restante <= 10:
                print(f"\n🕒 ¡ALERTA! Tiempo restante: {self.tiempo_restante} min. Regresando a casa inmediatamente.")
                break

            print(f"\n📍 Paseando por: {calle}...")
            time.sleep(1.5) # Simula el tiempo caminando
            self.tiempo_restante -= 7

            # Lógica de sensores durante el tramo
            evento = self.sensores_limpieza()

            if evento == 1: # Defecación
                if self.bolsitas > 0:
                    print("💩 El perro hizo popó. [ACCIÓN]: Sacando bolsita y recogiendo.")
                    self.bolsitas -= 1
                else:
                    print("❌ ERROR: Sin bolsitas. Notificando al dueño.")

            elif evento == 2: # Orina
                if self.tanque_agua_ml >= 200:
                    print("💦 El perro hizo pis. [ACCIÓN]: Lanzando chorro de agua desinfectante.")
                    self.tanque_agua_ml -= 200
                else:
                    print("💧 Depósito de agua bajo. Necesita recarga.")

            elif evento == 3: # Charco
                print("🔄 [OBSTÁCULO]: Charco detectado. Cambiando dirección para evitar lodo.")

        self.finalizar_recorrido()

    def finalizar_recorrido(self):
        print("\n--- RECORRIDO FINALIZADO ---")
        print(f"🏠 El robot ha llegado a casa con el perro.")
        print(f"📊 Reporte: Bolsitas restantes: {self.bolsitas} | Agua: {self.tanque_agua_ml}ml")

# --- INSTANCIAR Y CORRER EL ROBOT ---
mi_robot = RobotPaseador("DogBot-v1")
mi_robot.ejecutar_paseo()
```
